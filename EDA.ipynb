```python
import string 
from nltk.corpus import stopwords
```

```python
def process_text(sample_string):
    STOPWORDS=stopwords.words("english")
    no_punc=[i for i in sample_string if i not in string.punctuation]
    no_punc=''.join(no_punc)
    no_stop=[word for word in no_punc.split(" ") if word.lower not in STOPWORDS]
    return ' '.join(no_stop)
```

```python
review
```

```python
review["clean_text"]=review.Review.apply(process_text)
review
```

```python
from collections import Counter
words_counter=[]
words=[]
for i in range(5):
    words.append(review[review.Rating==i+1].clean_text.apply(lambda x : [word.lower() for word in x.split()]))
    words_counter.append(Counter())
    
    for word in words[i]:
        words_counter[i].update(word)
```

```python
print(words_counter[0].most_common(50))
```

```python
print(words_counter[4].most_common(50))
```

```python
for i in range(5):
    print("50 most common words for rating {}".format(i+1))
    print(words_counter[i].most_common(50))
```
